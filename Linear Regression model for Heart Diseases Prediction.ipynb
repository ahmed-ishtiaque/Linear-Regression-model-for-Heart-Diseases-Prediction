```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df=pd.read_csv("C:\\Users\\ishtiaque\\Desktop\\6b. ML Projects\\Linear Regression Model for heart diseases dataset\\heart.data.csv")
```

```python
df
```

```python
df.info()
```

```python
df=df.drop('Unnamed: 0',axis=1)
```

```python
X=df.drop('heart.disease',axis=1)
y=df['heart.disease']
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
scaler.fit(X)
X=scaler.transform(X)
y=y.to_numpy()
```

```python
sns.lmplot(x='biking',y='heart.disease',data=df)
```

```python
sns.lmplot(x='smoking',y='heart.disease',data=df)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
from sklearn import linear_model
clf=linear_model.LinearRegression()
```

```python
clf.fit(X_train,y_train)
```

```python
#R^2 value:
print(clf.score(X_train,y_train))
```

```python
y_pred=clf.predict(X_test)
print(y_test,y_pred)
print("Mean Squared Error:",np.mean(y_pred-y_test)**2)
```

```python
print(clf.coef_,clf.intercept_)
```
